```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python

df=pd.read_csv('city_day.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
x=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2','O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']
```

```python
y=['AQI_Bucket']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df[x],df[y],test_size=0.2)
```

```python
xtrain.shape,xtest.shape
```

```python
ytest.shape,ytrain.shape
```

```python
c=RandomForestClassifier()
c.fit(xtrain,ytrain)
```

```python
ypred=c.predict(xtest)
score=accuracy_score(ytest,ypred)
```

```python
cm=confusion_matrix(ytest,ypred)
print(cm)
```

```python
import pickle
pickle.dump(c,open('model.pkl','wb'))
```
